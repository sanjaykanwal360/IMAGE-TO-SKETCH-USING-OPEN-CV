```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("sample.jpg", 0)
### 0 is used as a flag
```

```python
plt.imshow(image,cmap='gray')
### Gives the output as greyscale image
```

```python
image.max()
### maximum value of that image
```

```python
image.min()
### min value of image 
```

```python
### To invert an image
Inv_image = 255 - image
```

```python
plt.imshow(Inv_image , cmap='gray')
```

Applying Gaussian Blur, we are removing the detailed information and the noise in image

```python
Blured_Image  = cv2.GaussianBlur(Inv_image , ksize =(21,21) , sigmaX=0 , sigmaY=0 )
### ksize is kernel size
```

```python
plt.imshow(Blured_Image , cmap ='gray')
```

## DODGING AND BURNING TECHNIQUE

Dodging and burning are terms used in photography for a technique used during the printing process to manipulate the exposure of a selected area(s) on a photographic print, deviating from the rest of the image's exposure.

```python
dodge = lambda img, mask: cv2.divide(img, 255-mask , scale = 256)
```

```python
Blended = dodge(image,Blured_Image)
```

```python
plt.imshow(Blended , cmap = 'gray' )
```

```python
cv2.imwrite("sample_sketch.jpg" , Blended )
```
